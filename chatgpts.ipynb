````python
# Libraries
import tkinter as tk
from tkinter import messagebox, filedialog
import PyPDF2
from tkinterdnd2 import TkinterDnD, DND_FILES
````

````python
# Class definition

class PDFMergerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Encadenar Archivos PDF")
        self.root.geometry("600x400")
        
        self.files = []  # Lista para almacenar los archivos PDF

        # Frame para los botones
        self.frame_buttons = tk.Frame(self.root)
        self.frame_buttons.pack(pady=10)

        # Botón para agregar archivos
        self.add_button = tk.Button(self.frame_buttons, text="Agregar Archivos", command=self.add_files)
        self.add_button.pack(side="left", padx=10)

        # Botón para fusionar los PDFs
        self.merge_button = tk.Button(self.frame_buttons, text="Fusionar PDFs", command=self.merge_pdfs)
        self.merge_button.pack(side="left", padx=10)

        # Lista de archivos (con capacidad de arrastrar y soltar)
        self.file_listbox = tk.Listbox(self.root, selectmode=tk.SINGLE, width=50, height=15)
        self.file_listbox.pack(pady=10)

        # Configurar drag-and-drop
        self.file_listbox.drop_target_register(DND_FILES)
        self.file_listbox.dnd_bind('<<Drop>>', self.on_drop)

    def add_files(self):
        """Agrega archivos manualmente mediante un diálogo de archivo"""
        file_paths = filedialog.askopenfilenames(filetypes=[("Archivos PDF", "*.pdf")])
        
        if file_paths:
            for file_path in file_paths:
                if file_path not in self.files:
                    self.files.append(file_path)
                    self.file_listbox.insert(tk.END, file_path)
                else:
                    messagebox.showwarning("Advertencia", f"El archivo {file_path} ya está en la lista.")

    def on_drop(self, event):
        """Añade archivos arrastrados a la lista"""
        file_path = event.data
        if file_path.endswith(".pdf"):
            if file_path not in self.files:
                self.files.append(file_path)
                self.file_listbox.insert(tk.END, file_path)
            else:
                messagebox.showwarning("Advertencia", f"El archivo {file_path} ya está en la lista.")
        else:
            messagebox.showwarning("Advertencia", "Solo puedes agregar archivos PDF.")

    def merge_pdfs(self):
        """Fusiona los PDFs en un único archivo"""
        if not self.files:
            messagebox.showwarning("Advertencia", "No hay archivos PDF para fusionar.")
            return
        
        # Crear un objeto PDF merger
        pdf_merger = PyPDF2.PdfMerger()
        
        # Agregar los archivos PDF en el orden que están en la lista
        for file in self.files:
            pdf_merger.append(file)
        
        # Guardar el archivo final
        save_path = filedialog.asksaveasfilename(defaultextension=".pdf", filetypes=[("PDF files", "*.pdf")])
        
        if save_path:
            pdf_merger.write(save_path)
            messagebox.showinfo("Éxito", f"PDF fusionado guardado en {save_path}")
            self.files.clear()
            self.file_listbox.delete(0, tk.END)

# # Inicializar la ventana de la aplicación
# if __name__ == "__main__":
#     root = TkinterDnD.Tk()  # Usamos TkinterDnD para soportar drag-and-drop
#     app = PDFMergerApp(root)
#     root.mainloop()
````

````python
## Main
root = TkinterDnD.Tk()  # Usamos TkinterDnD para soportar drag-and-drop
app = PDFMergerApp(root)
root.mainloop()
````

¡Claro! Para crear el programa que mencionas, se puede usar la librería `tkinter` para la interfaz gráfica y `PyPDF2` (o `pikepdf`) para trabajar con archivos PDF. Aquí te doy una solución con `tkinter` para la GUI y `PyPDF2` para fusionar los archivos PDF.

El flujo de trabajo será el siguiente:
1. La interfaz tendrá un panel donde puedes arrastrar y soltar archivos PDF.
2. Los archivos se mostrarán en una lista.
3. Puedes reorganizar el orden de los archivos en el panel.
4. Si agregas un archivo duplicado, aparecerá una advertencia.
5. Se podrá generar un PDF único con los archivos en el orden deseado.

### Descripción del Código:
1. **Interfaz gráfica:**
   - Utiliza `Tkinter` para crear una ventana con un `Listbox` donde puedes ver los archivos PDF.
   - Tiene botones para agregar archivos y para fusionarlos.
   - El `Listbox` permite arrastrar y soltar los archivos PDF.

2. **Agregando archivos:**
   - Puedes agregar archivos de dos maneras:
     - Usando el botón "Agregar Archivos" que abre un cuadro de diálogo para seleccionar archivos.
     - Arrastrando y soltando archivos en el `Listbox`.

3. **Verificación de duplicados:**
   - Si intentas agregar un archivo que ya está en la lista, muestra una advertencia.

4. **Fusionar archivos:**
   - Cuando presionas el botón "Fusionar PDFs", los archivos se combinan en un solo archivo PDF en el orden en que están en la lista.
   - El archivo fusionado se guarda en la ubicación que selecciones.

### 🛠️ Comando para crear el `.exe`

Guarda tu script en una carpeta (por ejemplo, `C:\ChainPDFs\`). Luego abre una terminal en esa carpeta y ejecuta:

```bash
pyinstaller --onefile --windowed --add-data "<ruta_a_tkinterdnd2>\tkdnd;tkinterdnd2\tkdnd" chain_PDFs_deep.py
```

> 🧠 Reemplaza `<ruta_a_tkinterdnd2>` por la ruta real donde se encuentra el paquete en tu entorno Python, por ejemplo:
>
> `C:\Users\TuUsuario\AppData\Local\Programs\Python\Python311\Lib\site-packages\tkinterdnd2`

El parámetro `--windowed` oculta la consola y `--add-data` es necesario porque `tkinterdnd2` necesita acceso a su carpeta `tkdnd`.

---

### 🗂️ Resultado

Una vez que termine el proceso, encontrarás tu `.exe` en la carpeta `dist/chain_PDFs_deep/`.
